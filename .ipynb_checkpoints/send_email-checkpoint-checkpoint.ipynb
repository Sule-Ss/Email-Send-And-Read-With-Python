#### Basit Eposta Gnderimi

```python
import ssl # hesaba ifreleme yapacak modl
import smtplib
```

```python
kullanici = "username@gmail.com"
sifre= "password"
```

```python
alici = kullanici
# baslik = 'Paython Gnderisi'
mesaj = "ilk mesaj"

context = ssl.create_default_context()
```

```python
port = 465   # hotmail ise smtp hotmailin port unu almalyz.
host = "smtp.gmail.com"

eposta_sunucu =  smtplib.SMTP_SSL(host= host, port = port, context = context)
eposta_sunucu.login(kullanici, sifre)
eposta_sunucu.sendmail(kullanici, alici, mesaj)
```

#### Eklentili Eposta Gnderimi

```python
from email.mime.multipart import MIMEMultipart

from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders # encode: stringe evirme

posta = MIMEMultipart() # obje
posta['From'] = kullanici
posta['To'] = Kullanici # oklu gnderimde [kullanici1, kullanici2,..]
posta['Subject'] = baslik

posta.attach(MIMEText(mesaj), 'plain')

eklenti_dosya_ismi = "random.jpg"

with(open(eklenti_dosya_ismi, 'rb')) as eklenti_dosyasi:
    payload = MIMEBase('application', 'octate-stream') # kullancyla sunucu arasndaki irtibatta mesaj objesi.ierisinde data tutar
    payload.set_payload((eklenti_dosyas).read())
    encoders.encode_base64(payload) # resim dosyasndan stringe evirme gibi dnlebilir
    
    payload.add_header('Content-Decomposition', 'attachment', filename = eklenti_dosya_ismi)
    posta.attach(payload)
    
    posta_str = posta.as_string()
    
```

```python
port = 465  
host = "smtp.gmail.com"

eposta_sunucu =  smtplib.SMTP_SSL(host= host, port = port, context = context)
eposta_sunucu.login(kullanici, sifre)
eposta_sunucu.sendmail(kullanici, alici, posta_str)
```
